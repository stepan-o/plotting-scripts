## Line styles

```python
color = 'cornflowerblue'
points = np.ones(5)  # Draw 5 points for each line
text_style = dict(horizontalalignment='right', verticalalignment='center',
                  fontsize=12, fontdict={'family': 'monospace'})


def format_axes(ax):
    ax.margins(0.2)
    ax.set_axis_off()


# Plot all line styles.
fig, ax = plt.subplots()

linestyles = ['-', '--', '-.', ':']
for y, linestyle in enumerate(linestyles):
    ax.text(-0.1, y, repr(linestyle), **text_style)
    ax.plot(y * points, linestyle=linestyle, color=color, linewidth=3)
    format_axes(ax)
    ax.set_title('line styles')

plt.show()
```

## The tale of two subplots



```python
x1 = np.linspace(0, 5)
x2 = np.linspace(0, 2)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, linewidth=1, linestyle=':', marker='o', markersize=2, color='mediumaquamarine')
plt.ylabel('Damped oscillation')

plt.title('A tale of 2 subplots')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, linewidth=2, linestyle='-', marker='s', markersize=3.5, color='tomato')
plt.ylabel('Undamped oscillation')

plt.show()
```

### A tale of 4 subplots -- Insurance company

An insurance company offers four plans, each of which had a price change this year.

**Plan A:** -10%

**Plan B:** +3%

**Plan C:** +8%

**Plan D:** +9%

<img src='img/stats/stat_mean_5.jpg' width=400>

```python
(-10 + 3 + 8 + 9) / 4
```

The mean is $\large{\frac{-10\% + 3\% + 8\% + 9\%}{4}=2.5\%}$

Under which situations would it be misleading to say that the average price of insurance at the company "went up by 2.5%"?

#### Plan A costs significantly less than the other plans.

```python
plans_cost = {'cost_old':[100, 1000, 1000, 1000], 'change_per':[0.9, 1.03, 1.08, 1.09]}
plans_cost_df = pd.DataFrame(plans_cost, index=['A', 'B', 'C', 'D'])
plans_cost_df['cost_new'] = plans_cost_df['cost_old'] * plans_cost_df['change_per']
plans_cost_df['change'] = plans_cost_df['cost_new'] - plans_cost_df['cost_old']
plans_cost_df['change_per'] = plans_cost_df['change'] / plans_cost_df['cost_old']
plans_cost_df['mean_change_per'] = plans_cost_df['change'].mean() / plans_cost_df['cost_old']
plans_cost_df
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4)

font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 12}

plt.rc('font', **font)

fig.suptitle('Cost of insurance plans', x=1.4, y=1)
fig.subplots_adjust(left=0.125, bottom=-0.1, right=2.6, top=0.8, wspace=0.5, hspace=0.2)

mean_change = plans_cost_df['change'].mean()
mean_change_per = round(plans_cost_df['change_per'].mean(), 3) * 100

axes[0].bar(plans_cost_df.index, plans_cost_df['change_per'] * 100)
axes[0].set_title('Average change of price is deceiving')
axes[0].set_ylabel('Change of cost, %')
axes[0].set_xlabel('Insurance plan')
axes[0].get_children()[0].set_color('gray')
axes[0].get_children()[0].set_edgecolor('k')
axes[0].get_children()[0].set_linewidth(0.5)
axes[0].text(0.05, -2.2, '{0}%'.format(plans_cost_df.loc['A', 'change_per'] * 100), color='k', rotation=90, ha='center', alpha=0.5)
axes[0].get_children()[1].set_color('lightgray')
axes[0].get_children()[1].set_edgecolor('k')
axes[0].get_children()[1].set_linewidth(0.5)
axes[0].text(1, 1.8, '{0}%'.format(plans_cost_df.loc['B', 'change_per'] * 100), color='k', rotation=90, ha='center', alpha=0.5)
axes[0].get_children()[2].set_color('darkgray')
axes[0].get_children()[2].set_edgecolor('k')
axes[0].get_children()[2].set_linewidth(0.5)
axes[0].text(2, 6.7, '{0}%'.format(plans_cost_df.loc['C', 'change_per'] * 100), color='k', rotation=90, ha='center', alpha=0.5)
axes[0].get_children()[3].set_color('darkgray')
axes[0].get_children()[3].set_edgecolor('k')
axes[0].get_children()[3].set_linewidth(0.5)
axes[0].text(3, 7.8, '{0}%'.format(plans_cost_df.loc['D', 'change_per'] * 100), color='k', rotation=90, ha='center', alpha=0.5)
axes[0].axhline(0, color='k')
axes[0].axhline(mean_change_per, linestyle='--', color='deeppink')
axes[0].text(-0.55, plans_cost_df['change_per'].mean() + 3.4, 'Mean change: {0}%'.format(mean_change_per), color='k')
axes[0].yaxis.grid(color='k', linewidth=0.5, linestyle=':')

x_loc = np.arange(4)
axes[1].bar(x_loc - 0.2, plans_cost_df['cost_old'], width=0.4, color='y', edgecolor='k', linewidth=0.5)
axes[1].bar(x_loc + 0.2, plans_cost_df['cost_new'], width=0.4, color='turquoise', edgecolor='k', linewidth=0.5)
axes[1].set_title('if one plan is much cheaper')
axes[1].set_facecolor('whitesmoke')
axes[1].set_ylabel('Cost of insurance plan, $')
axes[1].set_xlabel('Insurance plan')
axes[1].legend(["Old cost", "New cost"], loc='center left')
axes[1].yaxis.grid(color='darkgray', linewidth=1.5, linestyle=':')

axes[2].bar(plans_cost_df.index, plans_cost_df['change'])
axes[2].set_title("Other plans changed\na lot more than A")
axes[2].set_facecolor('darkblue')
axes[2].set_ylabel('Mean change, % of plan cost')
axes[2].set_xlabel('Insurance plan')
axes[2].get_children()[0].set_color('springgreen')
axes[2].get_children()[0].set_edgecolor('w')
axes[2].get_children()[0].set_linewidth(1.5)
axes[2].get_children()[1].set_color('yellow')
axes[2].get_children()[1].set_edgecolor('w')
axes[2].get_children()[1].set_linewidth(1.5)
axes[2].get_children()[2].set_color('gold')
axes[2].get_children()[2].set_edgecolor('w')
axes[2].get_children()[2].set_linewidth(1.5)
axes[2].get_children()[3].set_color('darkorange')
axes[2].get_children()[3].set_edgecolor('w')
axes[2].get_children()[3].set_linewidth(1.5)
axes[2].axhline(0, linestyle='-', linewidth=2, color='w')
axes[2].axhline(mean_change, linestyle='--', color='deeppink')
axes[2].text(-0.3, mean_change, 'Mean change:\n{0}$'.format(mean_change), color='white', va='center')
axes[2].yaxis.grid(color='lightgray', linewidth=1.3, linestyle=':')

axes[3].bar(plans_cost_df.index, round(plans_cost_df['mean_change_per'], 2) * 100)
axes[3].set_title('Mean change constitutes\nmuch bigger % of Plan A')
axes[3].get_children()[0].set_color('darkred')
axes[3].get_children()[0].set_edgecolor('red')
axes[3].get_children()[0].set_linewidth(2)
axes[3].get_children()[1].set_color('salmon')
axes[3].get_children()[1].set_edgecolor('k')
axes[3].get_children()[1].set_linewidth(0.5)
axes[3].get_children()[2].set_color('salmon')
axes[3].get_children()[2].set_edgecolor('k')
axes[3].get_children()[2].set_linewidth(0.5)
axes[3].get_children()[3].set_color('salmon')
axes[3].get_children()[3].set_edgecolor('k')
axes[3].get_children()[3].set_linewidth(0.5)
axes[3].set_ylabel('Mean change of cost, % of plan cost')
axes[3].set_xlabel('Insurance plan')
axes[3].legend(['Mean change of cost,\n%of plan cost'], loc='center right')
axes[3].text(0.5, 47.2, "Pinoccio's nose", rotation=-90, color='darkred')
axes[3].yaxis.grid(color='gray', linewidth=1, linestyle=':')

plt.show()
```

Assume that Plan A costs $\$$100 and the others plans cost $\$$1000. Then the actual price changes are -$\$$10, +$\$$30, +$\$$80, and +$\$$90, giving an average price change of +$\$$47.50, which is nearly 5% of the more expensive plans and nearly 50% of the smaller one. Obviously, grouping these plans together and averaging their changes as a percentage ignores meaningful difference between the plans.

In this case, Plan A is significantly cheaper than the other three plans; the price of the three has gone up by an average 6.67\%--more than double the reported 2.5\%. The average price of insurance among all four plans has gone up by 47.5\$. This represents 47.5% of the old price of Plan A, which itself has gone down by only 10%. This way, adding Plan A to the other three results in the lower perceived average price.

#### The plans have significantly different numbers of members.
Assume that 100 people use Plan A, 1,000 each use Plans B and C, and 10,000,000 use Plan D. In this case, nearly everyone experiences a 9% increase, and over 99.9% of the members had a change that was higher than the "2.5% average increase." Averaging these percentages together would not represent a meaningful analysis of the data.

## Axes Demo

Example use of plt.axes to create inset axes within the main plot axes.

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


# create some data to use for the plot
dt = 0.001
t = np.arange(0.0, 10.0, dt)
r = np.exp(-t[:1000] / 0.05)  # impulse response
x = np.random.randn(len(t))
s = np.convolve(x, r)[:len(x)] * dt  # colored noise

# the main axes is subplot(111) by default
plt.plot(t, s)
plt.axis([0, 1, 1.1 * np.min(s), 2 * np.max(s)])
plt.xlabel('time (s)')
plt.ylabel('current (nA)')
plt.title('Gaussian colored noise')

# this is an inset axes over the main axes
a = plt.axes([.65, .6, .2, .2], facecolor='k')
n, bins, patches = plt.hist(s, 400, density=True)
plt.title('Probability')
plt.xticks([])
plt.yticks([])

# this is another inset axes over the main axes
a = plt.axes([0.2, 0.6, .2, .2], facecolor='k')
plt.plot(t[:len(r)], r)
plt.title('Impulse response')
plt.xlim(0, 0.2)
plt.xticks([])
plt.yticks([])

plt.show()
```

## Pyplot scales

Create plots on different scales. Here a linear, a logarithmic, a symmetric logarithmic and a logit scale are shown. For further examples also see the [Scales](https://matplotlib.org/gallery/index.html#scales-examples) section of the gallery.

```python
from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

## Variable inspector

From [ipywidgets examples collection](https://github.com/jupyter-widgets/ipywidgets/blob/master/docs/source/examples/Variable%20Inspector.ipynb).

```python
import ipywidgets as widgets # Loads the Widget framework.
from IPython.core.magics.namespace import NamespaceMagics # Used to query namespace.

# For this example, hide these names, just to avoid polluting the namespace further
get_ipython().user_ns_hidden['widgets'] = widgets
get_ipython().user_ns_hidden['NamespaceMagics'] = NamespaceMagics
```

```python
class VariableInspectorWindow(object):
    instance = None
    
    def __init__(self, ipython):
        """Public constructor."""
        if VariableInspectorWindow.instance is not None:
            raise Exception("""Only one instance of the Variable Inspector can exist at a 
                time.  Call close() on the active instance before creating a new instance.
                If you have lost the handle to the active instance, you can re-obtain it
                via `VariableInspectorWindow.instance`.""")
        
        VariableInspectorWindow.instance = self
        self.closed = False
        self.namespace = NamespaceMagics()
        self.namespace.shell = ipython.kernel.shell
        
        self._box = widgets.Box()
        self._box.layout.overflow_y = 'scroll'
        self._table = widgets.HTML(value = 'Not hooked')
        self._box.children = [self._table]
        
        self._ipython = ipython
        self._ipython.events.register('post_run_cell', self._fill)
        
    def close(self):
        """Close and remove hooks."""
        if not self.closed:
            self._ipython.events.unregister('post_run_cell', self._fill)
            self._box.close()
            self.closed = True
            VariableInspectorWindow.instance = None

    def _fill(self):
        """Fill self with variable information."""
        values = self.namespace.who_ls()
        self._table.value = '<div class="rendered_html jp-RenderedHTMLCommon"><table><thead><tr><th>Name</th><th>Type</th><th>Value</th></tr></thead><tr><td>' + \
            '</td></tr><tr><td>'.join(['{0}</td><td>{1}</td><td>{2}'.format(v, type(eval(v)).__name__, str(eval(v))) for v in values]) + \
            '</td></tr></table></div>'

    def _ipython_display_(self):
        """Called when display() or pyout is used to display the Variable 
        Inspector."""
        self._box._ipython_display_()
```

```python
inspector = VariableInspectorWindow(get_ipython())
inspector
```
